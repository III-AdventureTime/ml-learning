````python
import numpy as np
import matplotlib.pyplot as plt
````

# matplotlib API
## Figures and Subplots 

````python
fig = plt.figure()  # create a blank figure

ax1 = fig.add_subplot(2, 3, 1)  # (rows, columns, position)
ax2 = fig.add_subplot(1, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.plot(np.random.randn(50).cumsum(), 'k--')  # will plot on the last subplot created
ax1.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))  # plot on `ax1`
ax2.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)  # plot on `ax2`

plt.show()
````

````python
fig, axes = plt.subplots(2, 2)  # create a figure (`fig`) with 2x2 subplots (axes[r,c])
# more arguments:
#   sharex, sharey
#   subplot_kw -- dict of keywords passed to add_subplot()

plt.show()
````

### spacing around subplots 

````python
fig, axes = plt.subplots(2, 2, sharex=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)

plt.subplots_adjust(wspace=0.5,  # ratio of horizontal spacing to plot width
                    hspace=0.15)  # ratio of vertical spacing to plot height
plt.show()
````

## Colors, Marker, and Line Styles

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(0, 4*np.pi, 40)
ax.plot(x, np.sin(x)*0.5,     linestyle='-',  color='g',                   label='1st')
ax.plot(x, np.cos(x*1.6)*0.7, linestyle='--', color='#ffaa00', marker='o', label='2nd')
ax.plot(x, np.cos(x*0.8),     linestyle=':',  color='#00aaaa', marker='+', label='3rd')

plt.legend(loc='best')  # add legend
plt.show()
````

## Ticks, Lables, and Legends  

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(0, 4*np.pi, 100)
ax.plot(x, np.sin(x)*0.5, linestyle='-')

ax.set_xlim([0, 8])  # set x limits of subplot `ax`
plt.ylim([-0.6, 0.6])  # set y limits of the *most recent created* subplot

print(ax.get_xlim())  # get x limits of subplot `ax`
print(plt.ylim())  #get y limits of the *most recent created* subplot

plt.show()
````

### title, axis labels, ticks, and ticklabels

````python
fig = plt.figure(figsize=(12,5))

#
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.random.randn(1000).cumsum())

ax1.set_xticks([0,250,500,750,1000])  # set x-tick positions
ax1.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                   rotation=30, fontsize='medium')  # set x-tick labels
ax1.set_xlabel('stages')  # set x-axis label
ax1.set_title('my matplotlib plot')  # set subplot title

#
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(np.random.randn(1000).cumsum())

ax2.set(title='another plot', xlabel='X', xlim=[0,1000], ylabel='Y')  # set properties

#
plt.subplots_adjust(wspace=0.25)
plt.show()
````

### legends 

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(0, 4*np.pi, 100)
ax.plot(x, np.sin(x*3)*0.3, label='1st')
ax.plot(x, np.cos(x)*0.7,   label='2nd')
ax.plot(x, np.cos(x*0.8),   label='3rd')

ax.legend(loc='best')  # add legend

plt.show()
````

## Annotations and Drawing 

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(0, 4*np.pi, 100)
ax.plot(x, np.sin(x*2)*np.sin(x*0.3+0.1))
ax.set(ylim=[-1.05,1.5])

# add text
ax.text(9, -0.8, 'some text', family='monospace', fontsize=12)

# add text and arrow
ax.annotate("maximum", 
            xy=(3.97,1.0), xytext=(3.97,1.3),  # arrow: from `xytext` to `xy`
            arrowprops=dict(facecolor='black', headwidth=4, width=1, headlength=4))

plt.show()
````

Drawing shapes:
* "Patches": objects representing common shapes.
* Some of these (like `Rectangle` and `Circle`) are found in `matplotlib.pyplot`, but the full set is in `matplotlib.patches`.

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# create shape objects
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.4)
circ = plt.Circle((0.7, 0.4), 0.15, color='#55aacc')
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='#33aa33', alpha=0.7)

# add to subplot
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

plt.show()
````

# Plotting with pandas and seaborn

## Line Plots

````python
# plotting pandas Series
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.name = 'position'
s.index.name = 't'

s.plot(style='--', title='random walk', xlim=(0,90))
plt.show()
````

````python
# plotting on an existing subplot
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

s.plot(ax=ax, style='g--o')
ax.set(title='random walk', xlim=(0,90), ylabel='position')

plt.show()
````

````python
# plotting pandas DataFrame
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.index.name = 't'
df.plot(xlim=(0,90), style='--o')

plt.show()
````

## Bar Plots

````python
# pandas Series
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.index.name = 'index'

fig, axes = plt.subplots(1, 2)
fig.set_size_inches(10, 3.5)
fig.subplots_adjust(wspace=0.25)

# vertical bar plot
data.plot.bar(ax=axes[0], color='k', alpha=0.5)
axes[0].set(ylabel='value')

# horizontal bar plot
data.plot.barh(ax=axes[1], color='k', alpha=0.5)
axes[1].set(xlabel='value')

plt.show()
````

````python
# pandas DataFrame
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df.index.name = 'phase'
print(df)

fig, axes = plt.subplots(1, 2)
fig.set_size_inches(14, 4)
fig.subplots_adjust(wspace=0.25)

df.plot.bar(ax=axes[0], alpha=0.6)
df.plot.barh(ax=axes[1], stacked=True, alpha=0.6)

plt.show()
````
